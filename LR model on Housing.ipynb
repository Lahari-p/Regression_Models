```python
import pandas as pd, numpy as np, matplotlib.pyplot as plt, seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\pedin\Downloads\Housing.csv")
df.head()
```

```python
df.isnull().mean()
```

```python
dum = pd.get_dummies(df[['mainroad','guestroom','basement','hotwaterheating','airconditioning','prefarea','furnishingstatus']],prefix=['mainroad','guestroom','basement','hotwaterheating','airconditioning','prefarea','furnishingstatus'],columns=['mainroad','guestroom','basement','hotwaterheating','airconditioning','prefarea','furnishingstatus'])
df1 = pd.concat([dum,df],axis=1)
df1.tail()
```

```python
df1.drop(['mainroad','guestroom','basement','hotwaterheating','airconditioning','prefarea','furnishingstatus'],axis=1,inplace=True)
df1.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test,y_train, y_test = train_test_split(df1.drop('price',axis=1),df1['price'],test_size=0.3,random_state=10)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.preprocessing import MinMaxScaler
sc = MinMaxScaler()
X_train = sc.fit_transform(X_train['area'])
```

```python
import statsmodels.api  as sm
X_train_sm = sm.add_constant(X_train)
model = sm.OLS(y_train, X_train_sm).fit()
model.summary()
```

```python
X_test = sc.transform(X_test['area'])
X_test_sm = sm.add_constant(X_test)
pred =model.predict(X_test_sm)
```
